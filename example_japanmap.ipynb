```python
import pandas as pd
from japanmap import picture
import matplotlib.pyplot as plt
```

```python
# read cvs
# data from https://www.nstac.go.jp/use/literacy/ssdse/
df = pd.read_csv('./.data/SSDSE-E-2024.csv', encoding='shift_jis', header=2)
print(df.columns)
df
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
target_col = "1人当たり県民所得（平成27年基準）"
plt.title('income in Japan')
plt.axis('off')
# カラーマップでmatplotlibのcmapのjetを設定
cmap = plt.get_cmap('jet')

# カラーマップの最小値と最大値を設定

norm = plt.Normalize(vmin=df[target_col].min(), vmax=df[target_col].max())
# カラーマップの凡例を表示
plt.colorbar(plt.cm.ScalarMappable(norm, cmap),ax=ax)

# データをSeries化する際に可視化対象のデータをカラーマップに従ってカラーコードに変換する関数を作成
fcol = lambda x: '#' + bytes(cmap(norm(x), bytes=True)[:3]).hex()
# japanmapライブラリとmatplotlibを使って可視化
plt.imshow(picture(df[target_col].apply(fcol)))
```
